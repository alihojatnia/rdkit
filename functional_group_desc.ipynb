```python
#return would be a list of number of functiolan groups
def check_for_smart(smiles,smarts):
  results_of_smarts = []
  mol = Chem.MolFromSmiles(smiles)
  
  for i in smarts:
    smart_pattern = Chem.MolFromSmarts(i)
    if mol.HasSubstructMatch(smart_pattern):
      results_of_smarts.append(len(mol.GetSubstructMatches(smart_pattern)))
    else:
      results_of_smarts.append(0)
  return results_of_smarts    

#list_to_series
def list_to_series(results_of_smarts:list,df:pd.DataFrame)->pd.Series: 
  return pd.Series(results_of_smarts, index=df.columns)
```

```python
#Smiles list
# sm = ['c1ccccc1Cl','CCCCCCCCCCCCCC','CCCCCCCCCCCCCCCCCCCC','CCCCCCCCCCCCCCCCCCCCCCCCCCCCCCC']

freesolv_df = pd.read_csv('https://deepchemdata.s3.us-west-1.amazonaws.com/datasets/freesolv.csv.gz')
sm = freesolv_df['smiles']

# Cols with the len(smarts)
cols = {str(i): 0 for i in range(len(smarts))}
# Create datafarame
df = pd.DataFrame(cols, index=[0])

#for loop over smiles to raech to the FGs desc
for i in range(len(sm)):
  df = df.append(list_to_series(check_for_smart(sm[i],smarts),df), ignore_index=True)

df = df.drop(0, axis=0)
df
```
