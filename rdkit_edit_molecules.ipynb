```python
from rdkit.Chem.Draw import IPythonConsole
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
mol = Chem.MolFromSmiles('ClCCCCCCCO')
mol
```

```python
changed_mol = Chem.DeleteSubstructs(mol,Chem.MolFromSmiles('Cl'))
changed_mol
```

```python
mol = Chem.MolFromSmiles('c1ccccc1Cl')
mol
```

```python
combinemol = Chem.CombineMols(changed_mol,mol)
combinemol
```

```python
mol = combinemol
for atom in mol.GetAtoms():
    atom.SetAtomMapNum(atom.GetIdx())
mol
```

```python
editable = Chem.EditableMol(combinemol)
editable.AddBond(2,11,order=Chem.rdchem.BondType.SINGLE)
editable.GetMol()
```

```python
editable = Chem.EditableMol(combinemol)
editable.AddBond(0,7,order=Chem.rdchem.BondType.DOUBLE)
editable.GetMol()
```

```python
m = Chem.MolFromSmiles('CCCC')
m = Chem.EditableMol(m)
m.AddAtom(Chem.Atom(8))
m.AddBond(0,4,order=Chem.rdchem.BondType.DOUBLE)
m.GetMol()
```

```python
Replace_mol = Chem.ReplaceSubstructs(Chem.MolFromSmiles('CCCCCO'), 
                                 Chem.MolFromSmiles('CO'), 
                                 Chem.MolFromSmarts('[#9]'),
                                 replaceAll=True)
Replace_mol[0]
```
